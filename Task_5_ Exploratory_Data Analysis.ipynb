```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('TITANIC DATASET.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Survived'].value_counts()
```

```python
# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)  # Too many missing values

# Convert categorical variables
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Histograms
df['Age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Boxplot
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
```

```python
# Pairplot
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived')

```

```python
# Survival by Sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')

# Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')

# Age vs Fare
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare (Survival)')
```


## Summary of EDA Findings

- **Age**: Younger passengers had slightly higher survival rates.
- **Gender**: Females were more likely to survive.
- **Class**: 1st class passengers had better survival chances.
- **Fare**: Higher fare generally associated with survival.
- **Embarked**: Port of embarkation may influence survival (C > S or Q).
